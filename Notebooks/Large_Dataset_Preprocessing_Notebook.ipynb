# Notebook to do the Preprocessing of the IMDB Dataset



# Import the Pandas module and read the title_basics file

```python
import pandas as pd
 

title_basics = pd.read_csv('title_basics.tsv', sep='\t')
 
title_basics
```

# Drop all rows with titleType="tvEpisode"

```python
title_basics = title_basics.drop(title_basics[title_basics.titleType == "tvEpisode"].index)
title_basics.head()
```

# Check the count of each type of value in titleType

```python
title_basics['titleType'].value_counts()
```

# Choose only the movie type of titleType and drop the rest

```python
title_basics = title_basics[title_basics['titleType'] == "movie"]
title_basics
```

# Remove blank/"\N" values from the dataset

```python
title_basics = title_basics.drop(title_basics[title_basics.genres == "\\N"].index)
title_basics
```

# Save csv of the dataset

```python
title_basics.to_csv("title_basics.csv", sep=',')
```

# Replace "\N" values with 0 in endYear attribute

```python
title_basics["endYear"].replace({"\\N": "0"}, inplace=True)
title_basics
```

# Replace "\N" values with 0 in runtimMinutes attribute

```python
title_basics["runtimeMinutes"].replace({"\\N": "0"}, inplace=True)
title_basics
```

# Replace "\N" values with 0 in startYear attribute

```python
title_basics["startYear"].replace({"\\N": "0"}, inplace=True)
title_basics
```

# Drop endYear attribute

```python
title_basics = title_basics.drop('endYear', 1)
title_basics
```

# Save csv of the titlebasics again

```python
title_basics.to_csv("title_basics_latest.csv", sep=',')
```

# Use one hot encoding on the dataset to split the genres

```python
list_here=[0]*(title_basics.shape[0])
title_basics['genres'].value_counts()
genres_list =title_basics['genres'].tolist()
for genre_here in genres_list:
    gl=genre_here.split(",")
    for genre in gl:
        if genre not in title_basics:
            title_basics[genre]=list_here
title_basics.head()
```

# Use one hot encoding on the dataset to set the values

```python
for index, row in title_basics.iterrows():
    gl=row['genres'].split(",")
    for genre in gl:
        title_basics.at[index,genre]=1
title_basics.head()
```

# Save latest CSV file

```python
title_basics.to_csv("title_basics_latest_1.csv", sep=',')
```

# Open previously saved CSV

```python
title_basics_latest = pd.read_csv('title_basics_latest_1.csv')
title_basics_latest
```

# Read title_ratings tsv file

```python
title_ratings = pd.read_csv('title_ratings.tsv', sep='\t')
 
# printing data
title_ratings
```

# Convert the "tconst" identifier to a list

```python
tconst_list_from_basics = title_basics_latest['tconst'].tolist()
```

# Use merge operation to combine the latest basics file and the ratings file

```python

final_data=pd.merge(title_basics_latest, title_ratings, on='tconst', how='outer')
final_data
```

# Remove the extra 'Unnamed: 0' column

```python
final_data = final_data.drop('Unnamed: 0', 1)
final_data
```

# Drop all rows with no title or titleType as NaN

```python
final_data = final_data.drop(final_data[final_data.titleType == "NaN"].index)
final_data
```

# Drop any remaining rows with NaN values in them

```python
final_data = final_data.dropna()
final_data
```

# Read the title crew tsv file.

```python
title_crew = pd.read_csv('title_crew.tsv', sep='\t')
title_crew
```

# Merge the final data and title crew data

```python
final_data_1=pd.merge(final_data, title_crew, on='tconst', how='inner')
final_data_1
```

# Remove "\N" values from the directors column

```python
final_data_1 = final_data_1.drop(final_data_1[final_data_1.directors == "\\N"].index)
final_data_1
```

# Remove "\N" values from the writers column

```python
final_data_1 = final_data_1.drop(final_data_1[final_data_1.writers == "\\N"].index)
final_data_1
```

# Drop the genres column as we have separated the genres using one hot encoding

```python
final_data_1 = final_data_1.drop('genres', 1)
final_data_1
```

# Save the dataset in csv as we will use it for further operations

```python
final_data_1.to_csv("final_data_for_algorithms.csv", sep=',')
```
